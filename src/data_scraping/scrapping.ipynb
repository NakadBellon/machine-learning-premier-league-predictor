```python
# Cell 1: Installation et imports

import soccerdata as sd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import os

print("✅ Packages installés et importés!")
```

```python
# Cell 2: Configuration du scraper FBref
class PremierLeagueScraper:
    def __init__(self):
        self.fbref = sd.FBref(leagues='ENG-Premier League')
        self.seasons = {
            '2019-2020': '2019-2020',
            '2020-2021': '2020-2021', 
            '2021-2022': '2021-2022',
            '2022-2023': '2022-2023',
            '2023-2024': '2023-2024',
            '2024-2025': '2024-2025',
            '2025-2026': '2025-2026'
        }
        
    def get_season_matches(self, season):
        """Récupère tous les matchs d'une saison"""
        print(f"📅 Scraping saison {season}...")
        try:
            matches = self.fbref.read_schedule(season)
            matches['season'] = season
            print(f"✅ {len(matches)} matchs trouvés")
            return matches
        except Exception as e:
            print(f"❌ Erreur {season}: {e}")
            return None
    
    def get_all_matches(self):
        """Récupère tous les matchs 2020-2026"""
        all_matches = []
        
        for season_name, season_code in self.seasons.items():
            season_matches = self.get_season_matches(season_code)
            if season_matches is not None:
                all_matches.append(season_matches)
        
        if all_matches:
            df = pd.concat(all_matches, ignore_index=True)
            print(f"\n🎉 TOTAL: {len(df)} matchs sur {len(self.seasons)} saisons")
            return df
        return None

# Initialisation
scraper = PremierLeagueScraper()
```

```python
# Cell 3: Scraping des matchs
print("🚀 Début du scraping Premier League 2020-2026...")

matches_df = scraper.get_all_matches()

if matches_df is not None:
    print(f"\n📊 Informations dataset:")
    print(f"Shape: {matches_df.shape}")
    print(f"Colonnes: {list(matches_df.columns)}")
    print(f"Saisons: {matches_df['season'].unique()}")
    
    # Aperçu des données
    print("\n👀 Aperçu des données:")
    display(matches_df.head())
else:
    print("❌ Aucune donnée récupérée")
```

```python
# Cell 4: Nettoyage des données
def clean_matches_data(df):
    """Nettoie et enrichit les données"""
    df = df.copy()
    
    # Parser le score
    df[['home_score', 'away_score']] = df['score'].str.split('–', expand=True)
    df['home_score'] = pd.to_numeric(df['home_score'], errors='coerce')
    df['away_score'] = pd.to_numeric(df['away_score'], errors='coerce')
    
    # Créer la variable cible (résultat)
    def get_result(row):
        if pd.isna(row['home_score']) or pd.isna(row['away_score']):
            return None
        if row['home_score'] > row['away_score']:
            return 'H'  # Home win
        elif row['home_score'] < row['away_score']:
            return 'A'  # Away win
        else:
            return 'D'  # Draw
    
    df['result'] = df.apply(get_result, axis=1)
    
    # Convertir la date
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    
    # Supprimer les matchs non joués
    df = df.dropna(subset=['home_score', 'away_score'])
    
    # Trier par date
    df = df.sort_values('date').reset_index(drop=True)
    
    print(f"✅ Données nettoyées: {len(df)} matchs joués")
    print(f"Distribution résultats: {df['result'].value_counts().to_dict()}")
    
    return df

matches_clean = clean_matches_data(matches_df)
```

```python
# Cell 5: Feature Engineering
def create_team_form_features(df, n_matches=5):
    """Crée des features basées sur la forme des équipes"""
    df = df.copy()
    
    # Initialiser les colonnes
    for prefix in ['home', 'away']:
        df[f'{prefix}_last_{n_matches}_points'] = 0.0
        df[f'{prefix}_last_{n_matches}_goals_scored'] = 0.0
        df[f'{prefix}_last_{n_matches}_goals_conceded'] = 0.0
        df[f'{prefix}_last_{n_matches}_xg'] = 0.0
    
    teams = pd.concat([df['home_team'], df['away_team']]).unique()
    
    for team in teams:
        # Matchs à domicile
        home_matches = df[df['home_team'] == team].index
        # Matchs à l'extérieur
        away_matches = df[df['away_team'] == team].index
        
        # Pour chaque match de l'équipe
        all_matches = sorted(list(home_matches) + list(away_matches))
        
        for i, match_idx in enumerate(all_matches):
            if i < n_matches:
                continue  # Pas assez d'historique
            
            # Récupérer les n derniers matchs
            last_matches_idx = all_matches[max(0, i-n_matches):i]
            
            # Calculer les stats
            points = 0
            goals_scored = 0
            goals_conceded = 0
            xg = 0
            
            for prev_idx in last_matches_idx:
                prev_match = df.loc[prev_idx]
                
                if prev_match['home_team'] == team:
                    goals_scored += prev_match['home_score']
                    goals_conceded += prev_match['away_score']
                    xg += prev_match['home_xg'] if pd.notna(prev_match['home_xg']) else 0
                    
                    if prev_match['result'] == 'H':
                        points += 3
                    elif prev_match['result'] == 'D':
                        points += 1
                else:
                    goals_scored += prev_match['away_score']
                    goals_conceded += prev_match['home_score']
                    xg += prev_match['away_xg'] if pd.notna(prev_match['away_xg']) else 0
                    
                    if prev_match['result'] == 'A':
                        points += 3
                    elif prev_match['result'] == 'D':
                        points += 1
            
            # Assigner les features
            if match_idx in home_matches:
                df.at[match_idx, f'home_last_{n_matches}_points'] = points
                df.at[match_idx, f'home_last_{n_matches}_goals_scored'] = goals_scored
                df.at[match_idx, f'home_last_{n_matches}_goals_conceded'] = goals_conceded
                df.at[match_idx, f'home_last_{n_matches}_xg'] = xg
            else:
                df.at[match_idx, f'away_last_{n_matches}_points'] = points
                df.at[match_idx, f'away_last_{n_matches}_goals_scored'] = goals_scored
                df.at[match_idx, f'away_last_{n_matches}_goals_conceded'] = goals_conceded
                df.at[match_idx, f'away_last_{n_matches}_xg'] = xg
    
    print(f"✅ Features de forme créées (derniers {n_matches} matchs)")
    return df

# Cette fonction peut prendre quelques minutes
print("⏳ Création des features... (patience!)")
matches_features = create_team_form_features(matches_clean, n_matches=5)
```
